# 30 Iterationen (Triplet)

```
import pandas as pd
url = 'https://raw.githubusercontent.com/LukasSchmauch/bin_packing_problem/main/results_30Iter_Unif_Trip.csv'
df = pd.read_csv(url)
df_grouped = df.groupby(['Typ', 'Anzahl Items'])
df_grouped.columns = ['Bin-Kapazitaet', ' Mean LB', 'Mean Hill Climbing', 'Mean First Fit Descending','Mean Abs. LB HC', 'Mean Abs. LB FFD', 'Mean Zeit HC (sec)', 'Mean Zeit FFD (sec)']
df_grouped.mean()
```
