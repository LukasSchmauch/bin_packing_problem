```
import pandas as pd
url = 'https://raw.githubusercontent.com/LukasSchmauch/bin_packing_problem/main/Results/results_unif_30.csv'
df = pd.read_csv(url)
df_grouped = df.groupby(['Typ', 'Anzahl Items'])
df_mean = df_grouped.mean()
df_mean.columns = ['Bin-Kapazitaet', ' Mean LB', 'Mean Hill Climbing', 'Mean First Fit Descending','Mean Abs. LB HC', 'Mean Abs. LB FFD', 'Mean Zeit HC (sec)', 'Mean Zeit FFD (sec)']
df_mean
```

# Verbesserung pro Iteration 501 Triplet Instanzen

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
url = 'https://raw.githubusercontent.com/LukasSchmauch/bin_packing_problem/main/Results/results_501_Itermessung1000.csv'
df = pd.read_csv(url)
grouped_df = df.groupby('Iteration').mean()
grouped_df.columns = ['Instanzindex', 'Anz_Items','Bin_Size','LB','HC-LB']

plt.figure(figsize=(14,10))
sns.lineplot(data=grouped_df["HC-LB"], label= 'Abweichung LB')
plt.title('Durchschnittliche Abweichung der HC L√∂sung von mittlerer LB pro Iteration', fontsize = 20)
plt.xlabel('Anzahl Iterationen', fontsize = 15)
plt.ylabel('Zus√§tzliche Bins gg√º. LB', fontsize = 15)
plt.yticks(np.arange(0, 24, 1))
plt.xticks(fontsize=18)
plt.yticks(fontsize=12)
plt.hlines(1,0,1000, colors='red', linestyles='solid', label = 'Abweichung = 1')
plt.legend()


```
